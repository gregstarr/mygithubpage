```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import *
import h5py
```

# Make an h5

```python
x = np.arange(-2,2,.2)
y = np.arange(-2,2,.2)
X,Y = np.meshgrid(x,y)
fxy = X*np.exp(-X**2 - Y**2)
Dx,Dy = np.gradient(fxy)

f = h5py.File("file.h5", 'w')
f.create_dataset("dx", data=Dx)
f.create_dataset("dy", data=Dy)
f.create_dataset("fxy", data=fxy)
f.close()
```

# Parameters

```python
nums = range(10)
items = [Text(value = "Box # "+str(i),disabled=True) for i in nums]
boxes = [Text() for i in nums]
HBox([VBox([i for i in items]),VBox([t for t in boxes])])
#enter name of file in box # 0
```
